```python
!curl -O http://owl.fish.washington.edu/halfshell/genomic-databank/Panopea-generosa-vv0.74.a3.exon.gff3 tracks/
```

```python
mv Panopea-generosa-vv0.74.a3.exon.gff3 tracks/Panopea-generosa-vv0.74.a3.exon.gff3
```

```python
!/Applications/bioinfo/bedtools2/bin/complementBed \
-i tracks/Panopea-generosa-vv0.74.a3.exon.gff3 \
-g analyses/Pg_genome.tab \
> analyses/Panopea-generosa-vv0.74.a3.exon.comp.bed
```

```python
!head analyses/Panopea-generosa-vv0.74.a3.exon.comp.bed
```

```python
!mv Panopea-generosa-vv0.74.a3.gene.gff3 tracks/Panopea-generosa-vv0.74.a3.gene.gff3   
```

```python
!/Applications/bioinfo/bedtools2/bin/intersectBed \
-a tracks/Panopea-generosa-vv0.74.a3.gene.gff3 \
-b analyses/Panopea-generosa-vv0.74.a3.exon.comp.bed \
| awk '{print $1, $4, $5}' > tracks/Panopea-generosa-vv0.74.a3.intron.bed
```
